```python
import numpy as np
from scipy import signal
from scipy import misc
import matplotlib.pyplot as plt
from PIL import Image
```

```python
gateImage = Image.open("Data_Training/IMG_0035.jpg")
#gateImage
```

```python
gScaleGateImage = gateImage.convert("L");
gScaleGateImage
```

```python
gScaleImageArray = np.asarray(gScaleGateImage)
gScaleImageArray.shape
#gScaleImageArray
```

```python
#Plotting
imgplot = plt.imshow(gScaleImageArray);
imgplot.set_cmap('gray')
plt.show(imgplot)
```

Implementing an edge detector kernel 

```python
kernel = np.array([[0,1,0],
                  [1,-4,1],
                  [0,1,0]])

edgeImage = signal.convolve2d(gScaleImageArray, kernel, mode='same', boundary='symm')
print(edgeImage.shape)
imgplot = plt.imshow(np.absolute(edgeImage))
imgplot.set_cmap("gray")
plt.show(imgplot)


```
